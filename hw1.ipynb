观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

```latex
DoSomething(A,p,r)
-----
n := r-p+1
if n=2 and A[p]>A[r] then
    swap A[p] and A[r]
else if n >= 3 then
    m = ceil(2n/3)
    DoSomething(A,p,p+m-1)
    DoSomething(A,r-m+1,r)
    DoSomething(A,p,p+m-1)  
    
---
first call: DoSomething(A,1,n)
```

note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. \* 使用python编程实现上述算法或其对应的更高效版本的算法

![问题2](./问题2.jpg)

算法功能:该算法的功能是对数组A进行升序排序。它不是最高效的解决方案。

````python
import math
def DoSomething(A, p, r):
    n = r - p + 1
    if n == 2:
        i = p - 1
        j = r - 1
        if A[i] > A[j]:
            A[i], A[j] = A[j], A[i]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        DoSomething(A, p, p + m - 1)
        DoSomething(A, r - m + 1, r)
        DoSomething(A, p, p + m - 1)

def sort_with_DoSomething(arr):
    if len(arr) <= 1:
        return arr
    DoSomething(arr, 1, len(arr))
    return arr

arr = [7,6,8,5, 4, 3, 2, 1]
print("原始:", arr)
sorted_arr = sort_with_DoSomething(arr)
print("排序:", sorted_arr)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

问题1：
该算法采用Kadane动态规划思想求解最大子数组和问题，通过一次遍历即可找到连续子数组的最大和及其起止位置。算法核心是维护两个变量：当前子数组和current_sum和全局最优解best_sum，遍历每个元素时，若当前子数组和加上新元素后小于新元素本身，则放弃之前子数组，从新元素位置重新开始；否则继续累加。每次更新后若current_sum大于best_sum，则记录新的最优解及其起止索引。
时间复杂度：O(n)

````python
def max_profit(profits):
    n = len(profits)
    if n == 0:
        return 0, None, None  
    best_sum = profits[0]
    current_sum = profits[0]
    start = 0
    end = 0
    temp_start = 0

    for i in range(1, n):
        if current_sum + profits[i] < profits[i]:

            current_sum = profits[i]
            temp_start = i
        else:
            current_sum += profits[i]

        if current_sum > best_sum:
            best_sum = current_sum
            start = temp_start
            end = i
    buy_day = start
    sell_day = end + 1

    return best_sum, buy_day, sell_day

profits =[3, 2, 1, -7, 5, 2, -1, 3, -1]
max_gain, buy_i, sell_i = max_profit(profits)
print("最大利润:", max_gain)
print("买入日:", buy_i+1, "卖出日:", sell_i)
print("实际收益验证:", sum(profits[buy_i:sell_i]))
````

1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

问题1：
算法设计思路：
因为 A[0] > A[1] 且 A[n-2] < A[n-1]，数组两端分别有下降和上升，所以中间必有局部最小。
使用用二分法：首先取中点 mid，如果 A[mid] < A[mid-1] 且 A[mid] < A[mid+1]，返回 A[mid]；否则，如果 A[mid] > A[mid+1]，说明右侧仍保持下降趋势，局部最小在右半段（[mid+1, r]。如果 A[mid] < A[mid+1]，说明左半有下降转向上升，局部最小在左半段（[l, mid]。这样可以保证 O(log n)。

idea:

````python
def minimum(A):
    n = len(A)
    l, r = 0, n - 1
    while l <= r:
        mid = (l + r) // 2
        left_val = A[mid - 1] if mid > 0 else float('inf')
        right_val = A[mid + 1] if mid < n - 1 else float('inf')
        if A[mid] < left_val and A[mid] < right_val:
            return A[mid] 
        if mid > 0 and A[mid - 1] < A[mid]:
            r = mid - 1
        else:
            l = mid + 1

A = [9, 3, 7, 2, 1, 4, 5]
print(minimum(A))
````

问题2：
算法设计思路：
直接扫描，对于每个位置 i，判断其是否比左右邻居都小。

````python
def minima(A):
    n = len(A)
    minima = []
    for i in range(n):
        left = A[i-1] if i > 0 else float('inf')
        right = A[i+1] if i < n-1 else float('inf')
        if A[i] < left and A[i] < right:
            minima.append(A[i])
    return minima

A = [9, 3, 7, 2, 1, 4, 5]
print(minima(A))
````

3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

问题1：
采用冒泡排序，时间复杂度：O(n²)

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

def bubble(arr, k):
    sorted_arr = bubble_sort(arr.copy())
    return sorted_arr[:k]

arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print("前{}个最小元素: {}".format(k, bubble(arr, k)))
````

问题2：
时间复杂度：O(nlogn+nlogk)

````python
import heapq
def sort(arr, k):
    heap = []
    for val in arr:
        heapq.heappush(heap, val)
    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))
    return result

arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
print(sort(arr, 3))
````

问题3：
时间复杂度：O(nlogk)

````python
import heapq

def sort(data_stream, k):
    heap = []
    for i, val in enumerate(data_stream):
        if i < k:
            heapq.heappush(heap, -val)
        else:
            if -heap[0] > val:
                heapq.heapreplace(heap, -val)
    result = sorted([-x for x in heap])
    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(sort(arr, k))
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

算法设计思路：
使用栈来实现：
从左到右遍历每个星球，如果栈为空或栈顶星球向右运动，就直接将当前星球入栈；只有当栈顶星球向右、当前星球向左时才会发生碰撞，此时不断比较两者质量，质量大的存活，质量相等则两者同时消失，直到当前星球被摧毁或栈顶不再满足碰撞条件，最后将存活的星球入栈。遍历结束后，栈中剩余的星球就是最终结果。
时间复杂度：O(n)

````python
def star(stars):
    stack = []
    for a in stars:
        while stack and stack[-1] > 0 and a < 0:
            if abs(stack[-1]) > abs(a):
                a = 0
            elif abs(stack[-1]) < abs(a):
                stack.pop()
                continue
            else: 
                a = 0 
            break
        if a != 0:
            stack.append(a)
    return stack

A = [23, -8, 9, -3, -7, 9, -23, 22]
B=[-3,-6,2,8,5,-8,9,-2,1]
print(star(A))
print(star(B))
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

算法设计思路：
这个问题可以用动态规划求解，定义状态 dp[i][w] 表示考虑前 i 个物品、背包承重为 w 时的最大价值。对于每个物品，决策是选或不选：不选时价值为 dp[i-1][w]，选时需要剩余容量 w-wi 且价值为 dp[i-1][w-wi] + vi，取两者最大值。按物品顺序填充二维表格，最终 dp[n][W] 即为最大价值。
时间复杂度：O(n×W)

````python
def bag(items, capacity):
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        weight, value = items[i - 1]
        for w in range(capacity + 1):
            if w >= weight:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weight] + value)
            else:
                dp[i][w] = dp[i - 1][w]
    return dp[n][capacity]

items = [(5,10), (4,40), (6,30), (3,50)]
capacity = 9
print("最大价值:", bag(items, capacity))
````

e.g.:
---

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

算法设计思路：
利用二叉搜索树中序遍历结果为严格递增序列这一性质。通过一次中序遍历来定位两个被错误交换的节点：维护一个指向前驱节点的指针，当遍历到当前节点时，如果前驱节点的值大于当前节点的值，就说明发现了逆序对，此时将第一个逆序对中的前驱节点记为第一个错误节点，将当前节点记为第二个错误节点；如果后续又发现第二个逆序对，则只需更新第二个错误节点为新的当前节点（因为两个错误节点交换后会导致中序遍历中出现两处逆序，相邻交换则只出现一处）。遍历完整棵树后，交换这两个错误节点的值即可恢复二叉搜索树。
时间复杂度为O(n)

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        self.first = None
        self.second = None
        self.prev = None
        
        def inorder(node):
            if not node:
                return
            
            inorder(node.left)
            
            if self.prev and self.prev.val > node.val:
                if not self.first:
                    self.first = self.prev
                self.second = node
            
            self.prev = node
            inorder(node.right)
        
        inorder(root)
        
        if self.first and self.second:
            self.first.val, self.second.val = self.second.val, self.first.val

def list_to_tree(arr):
    if not arr:
        return None
    nodes = [None if val is None else TreeNode(val) for val in arr]
    for i, node in enumerate(nodes):
        if node:
            left_idx = 2 * i + 1
            right_idx = 2 * i + 2
            if left_idx < len(nodes):
                node.left = nodes[left_idx]
            if right_idx < len(nodes):
                node.right = nodes[right_idx]
    return nodes[0]

def tree_to_list(root):
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result

root1 = list_to_tree([1,3,None,None,2])
print("输出:", end=" ")
Solution().recoverTree(root1)
print(tree_to_list(root1))
root2 = list_to_tree([3,1,4,None,None,2])
print("输出:", end=" ")
Solution().recoverTree(root2)
print(tree_to_list(root2))
````

给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。



<div style="text-align:center;width:800px;"> <img alt="" src="./fig/hw3q6.png"> </div>

请分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

算法设计思路：
使用双指针法：初始化左指针left指向数组起点、右指针right指向终点，同时维护左侧最大高度left_max和右侧最大高度right_max。每次比较左右指针指向的柱子高度，若左边高度小于右边，则处理左边：如果当前高度大于等于left_max则更新left_max，否则将left_max与当前高度的差值累加到雨水量中，随后左指针右移；若右边高度较小则对称处理右边。这样通过一次遍历即可计算出每个位置能接的雨水量。
时间复杂度：O(n)

````python
from typing import List
class Solution:
    def trap(self, height: List[int]) -> int:
        if not height:
            return 0
        
        left, right = 0, len(height) - 1
        left_max = right_max = 0
        water = 0
        
        while left < right:
            if height[left] < height[right]:
                if height[left] >= left_max:
                    left_max = height[left]
                else:
                    water += left_max - height[left]
                left += 1
            else:
                if height[right] >= right_max:
                    right_max = height[right]
                else:
                    water += right_max - height[right]
                right -= 1
        
        return water

sol = Solution()
height1 = [0,1,0,2,1,0,1,3,2,1,2,1]
height2 = [4,2,0,3,2,5]
print(sol.trap(height1))
print(sol.trap(height2))
````

 输入：root = [1,2,3], 输出：25

idea:

算法设计思路：
采用深度优先搜索（DFS）前序遍历二叉树，从根节点开始维护当前路径所表示的数字，每访问一个节点就将当前数字乘以10再加上该节点的值，当到达叶子节点（无左右子节点）时，将当前数字累加到总和中；若非叶子节点则递归遍历左右子树继续计算。这样通过一次遍历即可累加所有根到叶子路径所表示的数字之和。
时间复杂度：O(n)

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        def dfs(node, current_sum):
            if not node:
                return 0
            current_sum = current_sum * 10 + node.val
            if not node.left and not node.right:
                return current_sum
            return dfs(node.left, current_sum) + dfs(node.right, current_sum)
        return dfs(root, 0)


def build_tree_from_list(arr):
    if not arr:
        return None
    nodes = [None if val is None else TreeNode(val) for val in arr]
    for i, node in enumerate(nodes):
        if node:
            left_idx = 2 * i + 1
            right_idx = 2 * i + 2
            if left_idx < len(nodes):
                node.left = nodes[left_idx]
            if right_idx < len(nodes):
                node.right = nodes[right_idx]
    return nodes[0]



root1 = build_tree_from_list([1, 2, 3])
print(Solution().sumNumbers(root1))
root2 = build_tree_from_list([4, 9, 0, 5, 1])
print(Solution().sumNumbers(root2))
````
